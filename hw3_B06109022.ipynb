````python

class grad_fun:    
    def __init__(self, a, b, x, y, eta):
        self.a = a
        self.b = b
        self.x = x
        self.y = y
        self.eta = eta # 學習速率
        
        self.a_old = a
        self.b_old = b
        
        self.loss = None;
        
    # 損失函數
    def mse(self):
        sqr_err = ((self.a + self.b*self.x) - self.y) ** 2
        return np.mean(sqr_err)
    
    def GD(self):
        grad_a = 2 * np.mean((self.a + self.b*self.x - self.y))
        grad_b = 2 * np.mean((self.a + self.b*self.x - self.y) * self.x)
        return grad_a, grad_b

    def SGD(self):
        idx = np.random.randint(len(x))
        grad_b = 2 * (self.b*self.x[idx] + self.a - self.y[idx]) * (self.x[idx])
        grad_a = 2 * (self.b*self.x[idx] + self.a - self.y[idx]) * (1)
        return grad_a, grad_b

    def update(self):
        # 計算梯度
        grad_a, grad_b = self.SGD()
        
        # 梯度更新
        self.a_old = self.a
        self.b_old = self.b
        self.a = self.a_old - self.eta*grad_a
        self.b = self.b_old - self.eta*grad_b
        self.loss = self.mse
````

````python
def flower_dataset():
        np.random.seed(1)
        m = 400 # 樣本數
        N = int(m/2) # 每類樣本數
        D = 2 # 維度
        X = np.zeros((m,D))
        Y = np.zeros((m,1))
        a = 4 # 花長

        for j in range(2):
            ix = range(N * j, N * (j + 1))
            t = np.linspace(j * 3.12, (j + 1) * 3.12, N) + np.random.randn(N) * 0.2  # theta
            r = a * np.sin(4 * t) + np.random.randn(N) * 0.2  # radius
            X[ix] = np.c_[r * np.sin(t), r * np.cos(t)]
            Y[ix] = j

        X = X.T
        Y = Y.T

        return X, Y

X,Y = flower_dataset()
plt.scatter(X[0, :], X[1, :], c=Y, s=40, cmap=plt.cm.Spectral)
plt.show()
````

````python
def sigmoid(x):
    s = 1/(1+np.exp(-x))
    return s

# 定義神經網絡
def layer_size(X, Y):
    n_input =  X.shape[0]
    n_h = 4
    n_output = Y.shape[0]
    return (n_input, n_h, n_output)

# 初始化參數
def init_paramters(n_input, n_h, n_output):
    np.random.seed(2)
    W1 = np.random.randn(n_h, n_input) * 0.01
    b1 = np.zeros((n_h, 1))

    W2 = np.random.randn(n_output, n_h) * 0.01
    b2 = np.zeros((n_output, 1))

    parmaters={
        "W1":W1,
        "b1":b1,
        "W2":W2,
        "b2":b2
    }

    return parmaters
````

- 前向傳遞

\begin{align}
  z_1 &= W_1x + b_1\\
  a_1 &= {\rm tanh}(z_1)\\
  z_2 &= W_2x + b_2\\
  \hat{y} &= a_2 = \sigma(z_2)\\
  y &= \begin{cases}
    1, \text{if}\ a_2>0.5\\
    0, \text{o.w.}
  \end{cases}\\
  J &= \frac{−1}{m} \sum_{i=0}^m \big( y {\rm log}(a_2)+(1−y){\rm log}(1−a_2) \big)
\end{align}

> 請參考神經網路數學模型中的1~4式，完成`forward_prop`。 (5%)

````python
def forward_prop(X,params):
    W1,W2,b1,b2 = params['W1'],params['W2'],params['b1'],params['b2']

    Z1 =  np.dot(W1,X) + b1 
    A1 =  np.tanh(Z1)
    Z2 =  np.dot(W2,A1) + b2
    A2 =  sigmoid(Z2)

    #儲存結果，BP時使用
    cache={
        "Z1":Z1,
        "A1":A1,
        "Z2":Z2,
        "A2":A2
    }

    return A2, cache
````

````python
def compute_cost(A2, Y, params):
    m=Y.shape[1]
    J=np.sum(Y*np.log(A2)+(1-Y)*np.log(1-A2))/m
    cost=float(np.squeeze(J))
    return -cost
````

- 後向傳遞

>請用上課所教chain rule的方式，完成`backward_prop`。(20%)

````python
def backward_prop(params, cache, X, Y):
    A1,A2,W1,W2 = cache['A1'],cache['A2'],params['W1'],params['W2']
    m = Y.shape[1]

    dZ2= A2 - Y
    dW2 = 1 / m *(np.dot(dZ2,A1.T))
    db2 = np.sum(dZ2, axis=1, keepdims=True)/m
    dZ1 = np.dot(W2.T,dZ2) * (1 - np.power(A1, 2))
    dW1 = 1 / m *(np.dot(dZ1,X.T))
    db1 = np.sum(dZ1, axis=1, keepdims=True)/m

    grads = {
        "dW1":dW1,
        "dW2":dW2,
        "db1":db1,
        "db2":db2
        }

    return  grads
````

> 完成`update_params`。(5%)

````python
def updata_params(params, grads, lr=1.2):
    W1,W2,b1,b2 = params['W1'],params['W2'],params['b1'],params['b2']
    dW1,dW2,db1,db2 = grads['dW1'],grads['dW2'],grads['db1'],grads['db2']

    W1 = W1-lr*dW1
    W2 = W2-lr*dW2
    b1 = b1-lr*db1
    b2 = b2-lr*db2

    parmaters = {
        "W1": W1,
        "b1": b1,
        "W2": W2,
        "b2": b2
    }
    
    return parmaters
````

- 神經網路

````python
def nn_model(X, Y, n_h, iters):
    # X:訓練數據
    # Y:標籤
    # n_h:隱藏層節點數

    np.random.seed(3)
    n_input, n_h, n_output = layer_size(X, Y)
    params = init_paramters(n_input, n_h, n_output)
    W1,W2,b1,b2 = params['W1'],params['W2'],params['b1'],params['b2']
   
    for i in range(iters):
        A2, cache = forward_prop(X, params)

        grads = backward_prop(params, cache, X, Y)
        params = updata_params(params, grads)

        cost = compute_cost(A2, Y, params)

        if i%10000 == 0:
            print(f'iter={i}, cost={cost:.4f}')
            
    return params
````

>請參考第5式完成`predict`。(5%)

````python
def predict(X, params):
    A2, cache = forward_prop(X, params)
    predictions = np.round(A2)
    predictions = (A2 > 0.5)
    return predictions
````
